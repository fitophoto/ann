```python
from PIL import Image
import PIL
import sys
import os
```

```python
def process(im, rotate=True):
    if rotate:
        im = im.rotate(270, PIL.Image.NEAREST, expand = 1)
    return im.resize((800,800))
```

```python
dir_path="Photo/"
```

```python
for cls in os.listdir(dir_path):
    path = dir_path + cls + "/"
    out = f"Data800/{cls}/"
    for cat in os.listdir(path):
        back = path + cat + "/"
        for img in os.listdir(back):
            im = process(Image.open(back+img))
            if im.size != (800, 800):
                continue
            im.save(out+img)
```
