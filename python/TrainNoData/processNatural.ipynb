```python
import os
import pillow_heif
from PIL import Image
```

```python
path = "BackClassifierRaw/natural"
out = "BackClassifier/natural"
```

```python
def process(im, rotate=True):
    return im.resize((400,400))
```

```python
for file in os.listdir(path):
    heif_file = pillow_heif.read_heif("BackClassifierRaw/natural/"+file)
    image = Image.frombytes(
        heif_file.mode,
        heif_file.size,
        heif_file.data,
        "raw",
    )
    image = process(image)
    image.save(out+"/"+os.path.splitext(file)[0]+".jpg")
```
